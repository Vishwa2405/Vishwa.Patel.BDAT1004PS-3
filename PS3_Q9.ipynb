Create a pie chart showing the fraction of all Roman Emperors that were assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
file_path = '/content/roman-emperor-reigns.csv'
emperorsData = pd.read_csv(file_path)
```

```python
# New column 'Assassinated' that is True if the word 'Assassinated' is in the 'Cause_of_Death' column
emperorsData['Assassinated'] = emperorsData['Cause_of_Death'].str.contains('Assassinated', case=False, na=False)
```

```python
# Count the number of assassinated emperors & those who died from other causes
assassinationCounts = emperorsData['Assassinated'].value_counts()
```

```python
labels = ['Assassinated', 'Died of Other Causes']
sizes = assassinationCounts.values
explode = (0.1, 0)

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal')

plt.title('Proportion of Roman Emperors Assassinated')
plt.show()
```
