```python
import pandas as pd
```

```python
# Step 2 and 3: Import the dataset assign it to a variable called users
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users
```

```python
# Step 4. Discover what is the mean age per occupation
mean_age_per_occupation = users.groupby('occupation').age.mean()
print(mean_age_per_occupation)
```

```python
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least
def maleRatio(x):
    return x[x == 'M'].count() / x.count()

male_ratio_per_occupation = users.groupby('occupation').gender.apply(maleRatio).sort_values(ascending=False)
print(male_ratio_per_occupation)
```

```python
# Step 6. For each occupation, calculate the minimum and maximum ages
min_max_age_per_occupation = users.groupby('occupation').age.agg(['min', 'max'])
print(min_max_age_per_occupation)
```

```python
# Step 7. For each combination of occupation and sex, calculate the mean age
mean_age_per_occupation_and_sex = users.groupby(['occupation', 'gender']).age.mean()
print(mean_age_per_occupation_and_sex)
```

```python
# Step 8. For each occupation present the percentage of women and men
def percentageSex(x):
    return x.value_counts(normalize=True) * 100

percentage_sex_per_occupation = users.groupby('occupation').gender.apply(percentageSex).unstack()
print(percentage_sex_per_occupation)
```
