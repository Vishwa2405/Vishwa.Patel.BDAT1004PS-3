Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv('/content/us-marriages-divorces-1867-2014.csv')
data
```

```python
# Filter the data for the years 1900, 1950, and 2000
selectedYears = data[data['Year'].isin([1900, 1950, 2000])]
```

```python
# Plotting area
plt.figure(figsize=(10, 6))

width = 0.35
x = range(len(selectedYears['Year']))
plt.bar(x, selectedYears['Marriages_per_1000'], width, label='Marriages per 1000', color='blue')
plt.bar([p + width for p in x], selectedYears['Divorces_per_1000'], width, label='Divorces per 1000', color='red')

plt.title('Number of Marriages and Divorces Per 1000 people in the U.S. for Years 1900, 1950, and 2000')
plt.xlabel('Year')
plt.ylabel('per 1000 people')

# Setting the x-ticks
plt.xticks([p + width / 2 for p in x], selectedYears['Year'])
plt.legend()

# Visualization of the plot
plt.show()
```
