```python
# Step 1. Import the necessary libraries
import pandas as pd
```

```python
# Step 2 and 3. Import the dataset from this address and Assign it to a variable called chipo.
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
chipo
```

```python
# Step 4. See the first 10 entries
first_10_entries = chipo.head(10)
first_10_entries
```

```python
# Step 5: What is the number of observations in the dataset?
numObservations = chipo.shape[0]
print(numObservations)
```

```python
# Step 6: What is the number of columns in the dataset?
numColumns = chipo.shape[1]
print(numColumns)
```

```python
# Step 7: Print the name of all the columns.
columnNames = chipo.columns
print(columnNames)
```

```python
# Step 8: How is the dataset indexed?
datasetIndex = chipo.index
print(datasetIndex)
```

```python
# Step 9: Which was the most-ordered item?
most_ordered_item = chipo.groupby('item_name').quantity.sum().idxmax()
print(most_ordered_item)
```

```python
# Step 10: For the most-ordered item, how many items were ordered?
most_ordered_item_quantity = chipo.groupby('item_name').quantity.sum().max()
print(most_ordered_item_quantity)
```

```python
# Step 11: What was the most ordered item in the choice_description column?
most_ordered_choice = chipo.groupby('choice_description').quantity.sum().idxmax()
print(most_ordered_choice)
```

```python
# Step 12: How many items were ordered in total?
total_items_ordered = chipo.quantity.sum()
print(total_items_ordered)
```

```python
# Step 13: Turn the item price into a float and check the item price type
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x.replace('$', '')))
print(chipo['item_price'].dtype)
```

```python
# Step 14: How much was the revenue for the period in the dataset?
totalRevenue = (chipo['quantity'] * chipo['item_price']).sum()
print(totalRevenue)
```

```python
# Step 15: How many orders were made in the period?
totalOrders = chipo.order_id.nunique()
print(totalOrders)
```

```python
# Step 16: What is the average revenue amount per order?
average_revenue_per_order = totalRevenue / totalOrders
print(average_revenue_per_order)
```

```python
# Step 17: How many different items are sold?
num_different_items = chipo['item_name'].nunique()
print(num_different_items)
```
