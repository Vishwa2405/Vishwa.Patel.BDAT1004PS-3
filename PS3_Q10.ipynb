Question 10<BR>
Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009.
<BR>Don't forget to label your axes!
<BR>Color each dot according to its year.

```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
filePath = '/content/arcade-revenue-vs-cs-doctorates.csv'
data = pd.read_csv(filePath)
```

```python
# Checking the missing values
missingData = data.isnull().sum()
print("Missing data found:", missingData)
```

```python
if missingData.sum() == 0:
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(data['Total Arcade Revenue (billions)'],
                          data['Computer Science Doctorates Awarded (US)'],
                          c=data['Year'], cmap='viridis', s=100, edgecolors='black')

    plt.xlabel('Total Arcade Revenue (billions of USD)')
    plt.ylabel('Computer Science PhDs Awarded in the U.S.')
    plt.title('Relationship Between Arcade Revenue and CS PhDs Awarded (2000-2009)')
    plt.grid(True)

    cBar = plt.colorbar(scatter)
    cBar.set_label('Year')

    # Visualization of the plot
    plt.show()
else:
    print("Handle the missing data before plotting.")
```
