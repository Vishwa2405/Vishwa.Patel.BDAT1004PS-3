<br>Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name. Don't forget to label your axes!

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
filePath = '/content/actor_kill_counts.csv'
data = pd.read_csv(filePath)
```

```python
# Sort the data by kill count in ascending order so the highest counts are at the bottom of the chart
sorted_actor_data = data.sort_values('Count', ascending=True)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(5, 8))

barColor = plt.cm.tab20(range(len(sorted_actor_data)))
bars = plt.barh(sorted_actor_data['Actor'], sorted_actor_data['Count'], color=barColor)

plt.xlabel('Kill Count', fontsize=14, labelpad=10)
plt.ylabel('Actor', fontsize=14, labelpad=10)
plt.title('Deadliest Actors in Hollywood by Kill Count', fontsize=16)

for bar in bars:
    plt.text(
        bar.get_width() - 3,
        bar.get_y() + bar.get_height() / 2,
        f'{int(bar.get_width())}',
        va='center',
        ha='right',
        color='black',
        fontsize=8
    )

plt.tight_layout()
plt_path = '/content/deadliest_actors_improved_v2.png'
plt.savefig(plt_path)
plt.show()
```
